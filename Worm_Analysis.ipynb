```python
import glob
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import compare_ssim as ssim
%matplotlib inline
```

```python
# load image data for plate A2
file_list = glob.glob("/Users/yifei-wang/Desktop/Worm/Data/20190215/A2*.tif")
file_list.sort()
# show images 
fig = plt.figure(figsize=(20,20))
for n in range(len(file_list)):
    ax = fig.add_subplot(int(np.ceil(len(file_list)/5)),5,n+1)
    ax.imshow(plt.imread(file_list[n]),cmap='Greys_r')
# calculate structural similarity between two images (t,t+1)
SSIM = np.zeros(shape=(len(file_list)-1,1))
for n in range(len(file_list)-1):
    data1 = plt.imread(file_list[n])
    data2 = plt.imread(file_list[n+1])
    SSIM[n]= ssim(data1,data2)
# plot results
fig, ax = plt.subplots(figsize=(10, 8))
im = ax.scatter(range(len(file_list)-1), SSIM, color='blue', marker='o')
ax.set_xlabel('Time',fontsize=15)
ax.set_ylabel('Structural Similarity',fontsize=15)
ax.set_title('Structural Similarity over Time (Plate A2)',fontsize=20)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show() 
```
